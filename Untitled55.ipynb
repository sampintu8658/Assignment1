```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



# Plot using Seaborn and Matplotlib to differentiate the datasets
plt.figure(figsize=(10, 6))
df = get_dataset('census')
sample = synth.sample(n_samples = 32000)
# Create a barplot for both datasets
sns.lineplot(x='workclass', y='occupation', data=df, color='blue', label='df')
sns.lineplot(x='workclass', y='occupation', data=sample, color='red', label='sample')

# Adding labels and title
plt.title('Comparison of df and sample Datasets')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the legend to differentiate between datasets
plt.legend(title='Datasets')

# Show the plot
plt.show()
```

```python
combined = pd.concat([sample, df], keys=['sample', 'df']).reset_index(level=0).rename(columns={"level_0": "dataset"})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Age distribution
sns.kdeplot(data=combined, x='age', hue='dataset')
plt.title('Age Distribution in Sample and DF')
plt.show()

# Example: Hours-per-week by dataset
sns.boxplot(data=combined, x='dataset', y='hours-per-week')
plt.title('Hours per Week Comparison')
plt.show()
```

```python
sns.scatterplot(data=sample, x='education-num', y='hours-per-week', label='Sample')
sns.scatterplot(data=df, x='education-num', y='hours-per-week', label='DF')
plt.legend()
plt.title('Education vs. Hours-per-Week in Sample and DF')
plt.show()
```

```python
sns.countplot(data=combined, x='education', hue='dataset')
plt.title('Education Levels in Sample and DF')
plt.xticks(rotation=90)
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Example: Workclass
crosstab = pd.crosstab(combined['dataset'], combined['workclass'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print(f"Chi-Square Test for Workclass: p-value={p}")
```

```python
from scipy.stats import ttest_ind

# Example: Age
stat, p = ttest_ind(sample['age'], df['age'], nan_policy='omit')
print(f"T-Test for Age: p-value={p}")
```

```python
import seaborn as sns

sns.pairplot(combined, hue='dataset', vars=['age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss'])
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Combine and scale numerical data
numerical_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
scaled_data = StandardScaler().fit_transform(combined[numerical_cols])

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
combined['PCA1'], combined['PCA2'] = pca_result[:, 0], pca_result[:, 1]

sns.scatterplot(data=combined, x='PCA1', y='PCA2', hue='dataset')
plt.title('PCA Analysis of Sample and DF')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter rows where age == 19 in both datasets
rows_sample = sample[sample['age'] == 19]
rows_df = df[df['age'] == 19]

# Set up the figure and subplots for comparison
plt.figure(figsize=(16, 12))

# 1. Distribution of Workclass (Categorical Comparison)
plt.subplot(2, 2, 1)
sns.countplot(data=rows_sample, x='workclass', order=rows_sample['workclass'].value_counts().index, palette="Blues")
plt.title("Workclass Distribution (Sample)")
plt.xticks(rotation=45)
plt.xlabel("Workclass")
plt.ylabel("Count")

plt.subplot(2, 2, 2)
sns.countplot(data=rows_df, x='workclass', order=rows_df['workclass'].value_counts().index, palette="Greens")
plt.title("Workclass Distribution (DF)")
plt.xticks(rotation=45)
plt.xlabel("Workclass")
plt.ylabel("Count")

# 2. Distribution of `education` (Categorical Comparison)
plt.subplot(2, 2, 3)
sns.countplot(data=rows_sample, x='education', order=rows_sample['education'].value_counts().index, palette="Blues")
plt.title("Education Distribution (Sample)")
plt.xticks(rotation=45)
plt.xlabel("Education")
plt.ylabel("Count")

plt.subplot(2, 2, 4)
sns.countplot(data=rows_df, x='education', order=rows_df['education'].value_counts().index, palette="Greens")
plt.title("Education Distribution (DF)")
plt.xticks(rotation=45)
plt.xlabel("Education")
plt.ylabel("Count")

plt.tight_layout()
plt.show()

# 3. KDE for Numeric Columns - Compare `hours-per-week`
plt.figure(figsize=(10, 6))
sns.kdeplot(rows_sample['hours-per-week'], label='Sample', fill=True, color='blue', alpha=0.5)
sns.kdeplot(rows_df['hours-per-week'], label='DF', fill=True, color='green', alpha=0.5)
plt.title("Comparison of Hours Per Week (Age 19)")
plt.xlabel("Hours Per Week")
plt.ylabel("Density")
plt.legend()
plt.show()

# 4. Pie Chart for Relationship (Categorical Comparison)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
rows_sample['relationship'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Blues"))
plt.title("Relationship Distribution (Sample)")

plt.subplot(1, 2, 2)
rows_df['relationship'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Greens"))
plt.title("Relationship Distribution (DF)")

plt.tight_layout()
plt.show()

# 5. Comparison of `capital-gain` - KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(rows_sample['capital-gain'], label='Sample', fill=True, color='blue', alpha=0.5)
sns.kdeplot(rows_df['capital-gain'], label='DF', fill=True, color='green', alpha=0.5)
plt.title("Comparison of Capital Gain (Age 19)")
plt.xlabel("Capital Gain")
plt.ylabel("Density")
plt.legend()
plt.show()
```
